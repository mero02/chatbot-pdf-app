```python
# Importar librerías
import pandas as pd
import os
```

```python
# Ruta base
ruta_base = "/content/drive/MyDrive/Proyectos/ChatBot"

# Cargar los CSV
df_30 = pd.read_csv(os.path.join(ruta_base, "respuestas_30.csv"))
df_80 = pd.read_csv(os.path.join(ruta_base, "respuestas_80.csv"))
df_100 = pd.read_csv(os.path.join(ruta_base, "respuestas_100.csv"))
```

```python
# Agregar columna para identificar origen
df_30["origen"] = "30"
df_80["origen"] = "80"
df_100["origen"] = "100"

# Unir todo
df_total = pd.concat([df_30, df_80, df_100], ignore_index=True)

# Normalizar preguntas para agrupar
df_total["pregunta_norm"] = df_total["pregunta"].str.strip().str.lower()

# Agrupar por pregunta normalizada
agrupado = df_total.groupby("pregunta_norm")

# Lista para guardar resultados
filas_finales = []
```

```python
# Evaluación manual
for pregunta_norm, grupo in agrupado:
    print("="*80)
    pregunta_original = grupo.iloc[0]["pregunta"]
    tipo_pregunta = grupo.iloc[0]["tipo_pregunta"]
    respuesta_esperada = grupo.iloc[0]["respuesta_esperada"]

    print(f" Pregunta: {pregunta_original}")
    print(f" Respuesta esperada: {respuesta_esperada}\n")

    for idx, fila in grupo.iterrows():
        print(f" Origen: {fila['origen']}")
        print(f"   Respuesta obtenida: {fila['respuesta_obtenida']}")
        print(f"   Precisión original: {fila['precision']}")
        print("-" * 40)

    # Entrada manual
    nueva_precision = input(" Ingresá nueva precisión final: ").strip()
    nueva_observacion = input(" Ingresá observación final: ").strip()

    fila_resultante = {
        "tipo_pregunta": tipo_pregunta,
        "pregunta": pregunta_original,
        "respuesta_esperada": respuesta_esperada,
        "respuesta_obtenida": grupo.iloc[0]["respuesta_obtenida"],  # podrías elegir la mejor si querés
        "precision": nueva_precision,
        "observaciones": nueva_observacion
    }

    filas_finales.append(fila_resultante)

    print(" Respuesta guardada.\n")
```

```python
# Crear DataFrame final
df_resultado = pd.DataFrame(filas_finales)
```

```python
# Guardar en Drive
ruta_salida = os.path.join(ruta_base, "respuestas_unificadas_manual.csv")
df_resultado.to_csv(ruta_salida, index=False)

print(f"\n Proceso completo. Archivo guardado en:\n{ruta_salida}")
```
