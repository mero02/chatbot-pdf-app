```python
# Carga de CSV y conversión a Dataset
from datasets import Dataset
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Proyectos/ChatBot/desarrollo_preguntas_respuestas.csv")
dataset = Dataset.from_pandas(df)
dataset = dataset.train_test_split(test_size=0.1)
```

```python
# Carga de modelo base y tokenizer
from transformers import AutoTokenizer, AutoModelForCausalLM
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Aplicación de LoRA
from peft import LoraConfig, get_peft_model
lora_config = LoraConfig(
    r=8, lora_alpha=16,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    bias="none"
)
model = get_peft_model(model, lora_config)
```

```python
# Tokenización
def tokenize_function(examples):
    texts = [f"[INST] Eres un asistente experto en desarrollo web. Tema: {tema}. Subtema: {subtema}. Responde claramente: {pregunta} [/INST] {respuesta}{tokenizer.eos_token}"
             for tema, subtema, pregunta, respuesta in zip(
                 examples["Tema"], examples["Subtema"], examples["Pregunta"], examples["Respuesta"]
             )]
    tokenized = tokenizer(
        texts, truncation=True,
        max_length=128, padding="max_length",
        return_attention_mask=True
    )
    tokenized["labels"] = tokenized["input_ids"].copy()
    return tokenized

tokenized_dataset = dataset.map(tokenize_function, batched=True, batch_size=8)
```

```python
# Entrenamiento con Trainer
from transformers import TrainingArguments, Trainer
training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=4,
    gradient_accumulation_steps=8,
    logging_dir="./logs",
    save_total_limit=1,
    learning_rate=2e-4,
    fp16=True,
    num_train_epochs=3,
    save_steps=100,
    logging_steps=10,
    remove_unused_columns=False,
    label_names=["labels"]
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"]
)
```

```python
# Entrena el modelo
trainer.train()
```

```python
# Guarda el modelo local y en Drive
trainer.save_model("modelo_chatbot")
tokenizer.save_pretrained("modelo_chatbot")

from google.colab import drive
drive.mount('/content/drive')
model_path = "/content/drive/MyDrive/Proyectos/ChatBot/Modelo/modelo_chatbot"
trainer.save_model(model_path)
tokenizer.save_pretrained(model_path)
```
